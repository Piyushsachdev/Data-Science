```python
fig=plt.figure(figsize=(10,6))
ct=pd.crosstab(index=data1['Gender'],columns=data1['Caesarean'])
plt.bar(ct.index,ct['yes'],label='Born by Caesarean')
plt.bar(ct.index,ct['no'],bottom=ct['yes'],label='Not Born by Caesarean')
plt.xlabel('Gender')
plt.title('GENDER vs CAESAREAN',color='purple')
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(10,6))
ct=pd.crosstab(index=data1['Gender'],columns=data1['Smoke'])
plt.bar(ct.index,ct['yes'],label='do Smoke')
plt.bar(ct.index,ct['no'],bottom=ct['yes'],label=' dont do  Smoke')
plt.xlabel('Gender')
plt.title('GENDER vs Smoke',color='purple')
plt.legend()
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(20,15))
sns.violinplot(x='Age',data=data1,y='Smoke',hue='Caesarean',ax=axes[0,0])
axes[0,0].set_title('1. AGE vs SMOKE with respective Caesarean')
sns.violinplot(x='Gender',data=data1,y='Age',hue='Caesarean',ax=axes[0,1])
axes[0,1].set_title('2. GENDER vs AGE')
sns.swarmplot(y='LungCap',data=data1,x='Age',hue='Smoke',ax=axes[1,0])
axes[1,0].set_title('3. AGE vs LUNG CAPACITY')
sns.scatterplot(x='Height',data=data1,y='LungCap',hue='Smoke',ax=axes[1,1])
axes[1,1].set_title('4. HEIGHT vs LUNGCAPACITY')
plt.show()
```
